```python
bins = [0, 250, 500, 1000, 2000, df['Quantity'].max() + 1]
labels = ['0-250', '251-500', '501-1000', '1001-2000', '>2000']

# 2. Crear una columna que asigna cada transacci√≥n a un rango de volumen
df['Volume_Bin'] = pd.cut(
    df['Quantity'], 
    bins=bins, 
    labels=labels, 
    right=False, # Intervalos son [L√≠mite Inferior, L√≠mite Superior)
    include_lowest=True
)

# 3. Calcular el Precio Promedio y el Volumen Total por cada Rango (Bin)
df_analisis_descuento = df.groupby('Volume_Bin', observed=True).agg(
    Precio_Promedio=('Negotiated_Price', 'mean'),
    Total_Transacciones=('Negotiated_Price', 'count'),
    Volumen_Total=('Quantity', 'sum')
).reset_index()

# Ordenar por el Precio Promedio (de m√°s caro a m√°s barato) para confirmar el descuento
df_analisis_descuento = df_analisis_descuento.sort_values(by='Volume_Bin')

print("### üìä An√°lisis de Descuento por Volumen (Precio Promedio por Rango) ###")
print("-" * 70)
print(df_analisis_descuento)
print("-" * 70)
```

```python
L_F2 = {1: 0, 2: 1001, 3: 2001}
U_F2 = {1: 1000, 2: 2000, 3: np.inf}
K_F2 = L_F2.keys()

# 1. Funci√≥n para asignar cada transacci√≥n a su tramo de volumen (K)
def asignar_tramo_fase2(quantity):
    if quantity <= U_F2[1]:
        return 1
    elif quantity <= U_F2[2]:
        return 2
    else:
        return 3

# 2. Aplicar la funci√≥n al DataFrame para crear la columna 'Tramo_K_F2'
df['Tramo_K_F2'] = df['Quantity'].apply(asignar_tramo_fase2)

# 3. C√°lculo del Costo por Tramo C[i][j][k][t]
# Agrupaci√≥n por las 4 dimensiones (Item, Proveedor, Tramo, Periodo)
df_costo_tramo = df.groupby(['Item_Category', 'Supplier', 'Tramo_K_F2', 'Periodo'])['Negotiated_Price'].mean().reset_index(name='Costo_Tramo')

# 4. Conversi√≥n al Diccionario del Modelo
# Estructura: (Item, Proveedor, Tramo, Periodo) -> Costo
C_tramo_F2 = df_costo_tramo.set_index(['Item_Category', 'Supplier', 'Tramo_K_F2', 'Periodo'])['Costo_Tramo'].to_dict()

# 5. Impresi√≥n de Resultados (Muestra 5 elementos)
print("\n### Par√°metro de Costo por Tramo (C[i, j, k, t]) para la FASE 2 (L√≠mites corregidos) ###")
print(f"L√≠mites de los Tramos (K): 0-1000, 1001-2000, 2001+")
print("Diccionario C_tramo_F2: Muestra 5 elementos")
print(pd.Series(C_tramo_F2).head(5))
```

```python
FACTOR_IMPACTO = 0.50 # 50% del costo del producto para cubrir la falla

# --- 2. Preparaci√≥n del DataFrame para la Falla (Reutilizando la l√≥gica anterior) ---
# Creamos la columna binaria 'Falla' (1 si Defective_Units > 0, 0 si no).
if 'Defective_Units' in df.columns:
    df['Falla'] = np.where(df['Defective_Units'] > 0, 1, 0)
    print("‚úÖ Columna 'Falla' creada usando 'Defective_Units'.")
else:
    raise KeyError("Columna 'Defective_Units' no encontrada. No se puede calcular la Probabilidad de Falla.")

# --- 3. C√°lculo de la Probabilidad de Falla por Proveedor (Prob_j) ---
# Probabilidad = (√ìrdenes con Falla) / (Total de √ìrdenes) por Proveedor.
df_prob_falla = df.groupby('Supplier')['Falla'].mean().reset_index(name='Prob_Falla')

# --- 4. C√°lculo del Costo de Impacto PROMEDIO GLOBAL (50% del precio promedio general) ---
costo_promedio_general = df['Negotiated_Price'].mean()
costo_impacto_promedio = costo_promedio_general * FACTOR_IMPACTO

print(f"Costo Promedio Hist√≥rico General: ${costo_promedio_general:.2f}")
print(f"Costo de Impacto Promedio (50% del costo general): ${costo_impacto_promedio:.2f}")

# --- 5. C√°lculo de la Penalizaci√≥n por Riesgo Unitario (Pen_Riesgo_j) ---
# Penalizaci√≥n = Probabilidad de Falla * Costo de Impacto Promedio Global
df_prob_falla['Pen_Riesgo'] = df_prob_falla['Prob_Falla'] * costo_impacto_promedio

# --- 6. Conversi√≥n al Diccionario del Modelo ---
Pen_Riesgo_F3 = df_prob_falla.set_index('Supplier')['Pen_Riesgo'].to_dict()

# 7. Impresi√≥n de Resultados
print(f"\n### Par√°metro de Penalizaci√≥n por Riesgo (Pen_Riesgo_j) para la FASE 3 ###")
print(f"L√≥gica: Probabilidad de Falla x ${costo_impacto_promedio:.2f} (Costo Promedio Ajustado)")
print(Pen_Riesgo_F3)
```

```python
# =========================================================================
# 1. Definici√≥n del Par√°metro de Penalizaci√≥n por Riesgo (Pen_Riesgo_j)
# =========================================================================

# Costo de Impacto Ajustado (50% del precio promedio general)
COSTO_IMPACTO = df['Negotiated_Price'].mean() * 0.50

# Creaci√≥n de la columna 'Falla' (1 si hay unidades defectuosas, 0 si no)
# Asumimos que la columna es 'Defective_Units'
if 'Defective_Units' in df.columns:
    df['Falla'] = np.where(df['Defective_Units'] > 0, 1, 0)
else:
    # ‚ö†Ô∏è MANEJO DE ERROR: Si no existe, usamos una columna de prueba/simulaci√≥n
    # Si recibes un error, aseg√∫rate de que el nombre de la columna sea correcto
    print("‚ö†Ô∏è Columna 'Defective_Units' no encontrada. Se necesita el nombre correcto.")
    # Usaremos una lista vac√≠a para evitar fallas en la impresi√≥n
    Pen_Riesgo_j = {} 
    
# Si la columna Falla se cre√≥ correctamente, calculamos la penalizaci√≥n:
if 'Falla' in df.columns:
    # 1. Calcular Probabilidad de Falla por Proveedor (Prob_j)
    df_prob_falla = df.groupby('Supplier')['Falla'].mean().reset_index(name='Prob_Falla')
    
    # 2. Calcular Penalizaci√≥n
    df_prob_falla['Pen_Riesgo'] = df_prob_falla['Prob_Falla'] * COSTO_IMPACTO
    
    # 3. Convertir a diccionario
    Pen_Riesgo_j = df_prob_falla.set_index('Supplier')['Pen_Riesgo'].to_dict()

print("‚úÖ Par√°metro Pen_Riesgo_j (Penalizaci√≥n por Riesgo) extra√≠do.")
print("--- Muestra del diccionario: ---")
print(Pen_Riesgo_j)
```

```python
# =========================================================================
# 2. Definici√≥n Manual de L√≠mites de Tramos K
# =========================================================================
# Usamos np.inf (infinito) para el l√≠mite superior del √∫ltimo tramo.

L_K = {1: 0, 2: 1001, 3: 2001}
U_K = {1: 1000, 2: 2000, 3: np.inf}

print("\n‚úÖ L√≠mites L_K y U_K definidos manualmente.")
print(f"L√≠mites Inferiores (L_K): {L_K}")
print(f"L√≠mites Superiores (U_K): {U_K}")
```

```python

# =========================================================================
# 1. Creaci√≥n de la Columna 'Tramo_K'
# =========================================================================

# Definici√≥n de los l√≠mites de los tramos (breakpoints)
bins = [0, 1000, 2000, np.inf]
labels = [1, 2, 3] # K = {1, 2, 3}

# Crear la columna 'Tramo_K' en el DataFrame 'df'
# Esto clasifica cada fila (cada compra) en su tramo de descuento correspondiente.
df['Tramo_K'] = pd.cut(
    df['Quantity'],
    bins=bins,
    labels=labels,
    right=True,
    include_lowest=True
)

print("‚úÖ Columna 'Tramo_K' creada autom√°ticamente.")

# =========================================================================
# 2. Extracci√≥n del Par√°metro C_i_j_k_t
# =========================================================================

# Agrupaci√≥n por (Item, Proveedor, Tramo_K, Periodo) y c√°lculo del PROMEDIO de Negotiated_Price.
C_i_j_k_t = df.groupby(['Item_Category', 'Supplier', 'Tramo_K', 'Periodo'])['Negotiated_Price'].mean().to_dict()

print("‚úÖ Par√°metro C_i_j_k_t (Costo por Tramo) extra√≠do autom√°ticamente.")
print("--- Todos los costos por tramos ")
print(list(C_i_j_k_t.items()))
```
